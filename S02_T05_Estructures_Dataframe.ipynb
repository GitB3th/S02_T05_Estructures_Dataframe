```python
count_cancelled = sum(data_frame_2.Cancelled)
count_desviats = sum(data_frame.Diverted)
total_count=(count_cancelled + count_desviats)
print("There are a number of Flights Cancelled plus Desviats de: ", total_count)
```

```python
# Crear columnes noves
# velocitat mitjana de vol(Flight_Speed) --> miles/hour
# comprovant les dades,la distància ens ve donada en miles i AirTime en minuts(dividim entre 60 per passar a hores)
data_frame_2["Flight_Speed"] = ((data_frame_2["Distance"])/(data_frame_2["AirTime"]/60))

data_frame_2.head(5)
```

```python
data_frame_2.Flight_Speed.describe().round(2)
```

Observem que tenim valors infinits(inf) en la nova columna Flight_Speed, això és degut a alguns valors de la columna AirTime que són 0, i al dividir per 0 obtenim valor infinit
Verifiquem això!

```python
data_frame_2.AirTime.describe().round(2)
```

```python
#comprovem les files que tenen AirTime=0, q són les q ens provoquen els valors infinits
data_frame_2.loc[data_frame_2.AirTime == 0]
#passem aquests valors de inf a NaN values
data_frame_2.loc[data_frame_2['Flight_Speed'] == 'inf', 'Flight_Speed'] = 'nan'
#i comprovem
data_frame_2.AirTime.describe().round(2)
data_frame_2.loc[data_frame_2.AirTime == 0]
```

```python
#una altra opció de pandas es considerar els valors infinits com a NaN
pd.set_option('mode.use_inf_as_na', True)
```

```python
#columna nova si el vol ha arribat tard o no
#comprovem resum estadistic de la dada Arr.Delay
data_frame_2.ArrDelay.describe().round()
```

```python
Delay = []
for ArrDelay in data_frame_2["ArrDelay"]:
    if ArrDelay > 0.0:
        Delay.append(True)
    else:
        Delay.append(False)
data_frame_2["Delay"]=Delay

data_frame_2.head()
```

```python
# Taula de les aerolínies amb més endarreriments acumulats
data_frame_2.groupby("UniqueCarrier").size()
```

```python
data_frame_2.groupby("UniqueCarrier")["ArrDelay"].max()
```

```python
sorted(data_frame_2.groupby("UniqueCarrier")["ArrDelay"].max())
```

```python
#assignem el grup a una nova variable, i fem un plot, per veure millor les dades:
dades_delaybyCarrier= data_frame_2.groupby("UniqueCarrier")["ArrDelay"].max()
dades_delaybyCarrier.plot.bar()
#podem observar que el que més delay té NW (NorthWest Airlines), i el que menys delay és AQ (Aloha Airlines)
#M'he basat només en el parámetre ArrDelay, ja que he suposat que al final el que compte, és el temps que arriba respecte al programat!
#es podia haver fet la suma de tots els delays, però al final de tot, per part de l'usuari té una hora d'arribada prevista i li afecta arribar més tard de la hora
```

```python
#Quins són els vols més llargs?
pd.options.display.max_rows=None
vols_llargs = data_frame_2.groupby("FlightNum")["AirTime"].max()
vols_llargs = vols_llargs.sort_values(ascending=False)
vols_llargs.head()
```

```python
 #Quins són els més endarrerits?
vols_moredelayed = data_frame_2.groupby("FlightNum")["ArrDelay"].max()
vols_moredelayed = vols_moredelayed.sort_values(ascending=False)
vols_moredelayed.head() 
```
